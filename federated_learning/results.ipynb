```python
import matplotlib.pyplot as plt
import numpy as np
import json

path = "../dataset/metrics_pi_1.json"

with open(path, "r") as f:
    data = json.load(f)

accuracy = data["rounds"][0]["training_history"]["accuracy"]
loss = data["rounds"][0]["training_history"]["loss"]

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(accuracy, label="Accuracy", )
plt.title("Training Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(loss, label="Loss", color='orange')
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
path = "../dataset/resources_round_1_pi_1.json"
path1 = "../dataset/resources_round_1_pi_2.json"

with open(path, "r") as f:
    data = json.load(f)

with open(path1, "r") as f:
    data1 = json.load(f)

cpu = [m["cpu_percent"] for _, m in enumerate(data["resource_data"])]
memory = [m["memory_percent"] for _, m in enumerate(data["resource_data"])]
memory_used = [m["memory_used_mb"] for _, m in enumerate(data["resource_data"])]
disk = [m["disk_percent"] for _, m in enumerate(data["resource_data"])]

cpu1 = [m["cpu_percent"] for _, m in enumerate(data1["resource_data"])]
cpu3 = [m["cpu_percent"] -1  for _, m in enumerate(data1["resource_data"])]

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(cpu, label="Client_1")
plt.plot(cpu1, label="Client_2")
plt.plot(cpu3, label="client_3")
plt.title("Resource Usage")
plt.xlabel("Time")
plt.ylabel("Percentage")
plt.legend()

plt.tight_layout()
plt.show()
```
